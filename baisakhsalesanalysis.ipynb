```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_excel(r'C:\Users\Anusha\Downloads\Sales Book Report  main.xlsx')
df.head()
```

```python
df.dropna(subset=['DATE','CUSTOMER NAME'],inplace=True)
df.head()
```

```python
df.drop(columns=['MITI','REF NO','RATE','NON TAXABLE','TAXABLE','REMARKS','DIVISION','ENTRY BY','ENTRY TIME'],inplace=True)
df
```

```python
df[['CUSTOMER NAME']].value_counts()    #gives us the customer name with their counts of repetition or counts of purchase.
```

```python
df[['CUSTOMER NAME']].value_counts()[0:10]   #gives us the  top ten customer name with their purchase counts in descending order.
```

```python
plt.figure(figsize=(10,8))
plt.bar(list(df['CUSTOMER NAME'].value_counts()[0:5].keys()),list(df['CUSTOMER NAME'].value_counts()[0:5]),color=["red","blue","green","yellow","purple"])
#Gives us the bar plot oof the top 5 customer with their purchase counts.
plt.title("TOP 5 CUSTOMER WITH THEIR PURCHASE COUNT FOR BAISAKH")
```

```python
plt.figure(figsize=(8,8))
plt.bar(list(df['TRN MODE'].value_counts().keys()), list(df['TRN MODE'].value_counts()),color=["red","blue","green","black"])
plt.title("COUNTS OF TRANSACTION MODE")
```

```python
df.columns
```

```python
see = df[['DATE','BILL NO','CUSTOMER NAME','TRN MODE','GROSS AMOUNT','DISCOUNT','NET SALES']]
see  #we see the NET SALES DISCOUNT GROSS VALUE TRANCATION MODE CUSTOMER NAME AND BILL NUMBER WITH REFERENCE TO DATE
```

```python
sees = see.sort_values(by=['DISCOUNT'],ascending=False)
sees.head(10)    #wee see the columns of a dataframe with descending order with respect to discount, hence we see the maximum discount followed in descending order.
```

```python
plt.figure(figsize=(15,8))
plt.bar(list(sees['BILL NO'])[0:5], list(sees['DISCOUNT'])[0:5], color=["red","blue", "yellow","black","purple"])
plt.title("BILL NUMBER WITH DISCOUNT")
plt.xlabel("BILL NO")
plt.ylabel("DISCOUNT")
#visualizes the Bill number with discount i.e. which bill number gets what diiscount on descending order.
```

```python
plt.figure(figsize=(8,8))
plt.bar(list(sees['GROSS AMOUNT'])[0:5], list(sees['DISCOUNT'])[0:5],color=["red","blue","green","black"])
plt.title("GROSS VALUE WITH DISCOUNT")
plt.xlabel("GROSS AMOUNT")
plt.ylabel("DISCOUNT")
#visualizes the Net Sales with Discount i.e which GROSS VALUE GETS WHAT DISCOUNT
```
